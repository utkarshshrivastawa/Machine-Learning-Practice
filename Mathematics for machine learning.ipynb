# Linear Algebra 
1. Represents data: matrices for datasets, vectors for observations.
2. Enables operations: addition, multiplication, scaling, transformation.
3. Supports dimensionality reduction: PCA, SVD.
4. Facilitates neural networks: weights, activations.
5. Aids optimization: gradient descent, eigenvectors.

## Vector have Magnitude and Direction
1. Two vector lies in same direction but they have different magnitude
2. Vector is an array of value
3. Encodes features: numerical representation of data attributes for model input.
4. Facilitates computations: dot product, similarity, distance measures.
5. Supports algorithms: SVM, k-NN, clustering.

### Vector Operations 
1. Addition
2. Subtraction
3. Scalar Multiplication
4. Angle Calculation :
    1. If the angle between 2 vector is small, then the 2 vector are similar
    2. If the angle between 2 vector is large, then the 2 vector are very different.
5. Dot Product :The dot product of two vectors is a single number obtained by multiplying corresponding components and summing those products. It measures vector alignment.The scalar product of two vectors a and b of magnitude |a| and |b| is given as |a||b| cos θ, where θ represents the angle between the vectors a and b taken in the direction of the vectors.
6. Cross Product :If you have two vectors a and b then the vector product of a and b is c. So this a × b actually means that the magnitude of c = ab sinθ where θ is the angle between a and b and the direction of c is perpendicular to a well as b.
7. Vector Norm
8. Unit Vector
9. Projection :The projection vector formula in vector algebra for the projection of vector a on vector b is equal to the dot product of vector a and vector b, divided by the magnitude of vector b. The resultant of the dot product is a scalar value, and the magnitude of vector b is also a scalar value.
10. Linear Combination

### Projection of Vector

```python
a = np.array([2,5])
b = np.array([8,-6])
# magnitude of 'b' vector 
magnitude_of_b = np.sqrt(np.sum(b**2))
proj_of_a_on_b = (np.dot(a,b)/magnitude_of_b**2)*b
print("Projection of 'a' vector on the 'b' vector :",proj_of_a_on_b)
```

```python
a = np.array([2,3,5])
b = np.array([5,8,9])
# magnitude of 'b' vector 
magnitude_of_b = np.sqrt(np.sum(b**2))
proj_of_a_on_b = (np.dot(a,b)/magnitude_of_b**2)*b
print("Projection of 'a' vector on the 'b' vector:",proj_of_a_on_b)
```

### Matrix Addition

```python
a = np.array([[1,2,3],[5,4,6],[8,7,9]])
print(a)
b = np.array([[1,5,6],[5,8,9],[8,9,10]])
print(b)
# sum of two matrix is 
c= a+b # you cannot add the matrix with dissimilar shape
print("the sum of two matrix is : \n", c)
```

```python
# other way for addition 
a = np.array([[1,2,3],[5,4,6],[8,7,9]])
print(a)
b = np.array([[1,5,6],[5,8,9],[8,9,10]])
print(b)
# sum of two matrix is 
c= np.add(a,b)  # you cannot subtract the matrix with dissimilar shape
print("the sum of two matrix is : \n", c)
```

### Matrix Subtraction 

```python
a = np.array([[1,2,3],[5,4,6],[8,7,9]])
print(a)
b = np.array([[1,5,6],[5,8,9],[8,9,10]])
print(b)
# the substraction of 2 matrix is
c= a-b
print("The substraction of two matrix is :\n",c)
```

```python
# other way for subtraction
a = np.array([[1,2,3],[5,4,6],[8,7,9]])
print(a)
b = np.array([[1,5,6],[5,8,9],[8,9,10]])
print(b)
# the substraction of 2 matrix is
c= np.subtract(a,b)
print("The substraction of two matrix is :\n",c)

```
