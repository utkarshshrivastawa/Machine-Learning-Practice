#### Central Tendencies in Data Pre-Processing
Central Tendencies are very useful in handling the missing values in a dataset
1. Mean : Missing values in a dataset can be replaced with mean value, if the data is uniformly distributed.
2. Median: Missing values in a dataset can be replaced with median value, if the data is skewed.
3. Mode : Missing values in a dataset can be replaced with mode value, if the data is skewed. Missing categorical values can also be replaced with mode value.

### Measure of Variability
1. Range: The range of a set of data is the difference between the largest and smallest values. It can give a rough idea about the distribution of our dataset.
2. Variance: Variance is a measure of how far each number in the set is from the mean and therefore from every other number in the dataset.
3. Standard Deviation: Standard Deviation is the square root of Variance. Standard deviation looks at how spread out a group of numbers is from the mean.

#### Hypothesis
 The hypothesis is defined as the supposition or proposed explanation based on insufficient evidence or assumptions. It is just a guess based on some known facts but has not yet been proven. A good hypothesis is testable, which results in either true or false.
##### Types of Hypothesis
1. Null Hypothesis(H0): A null hypothesis is a type of statistical hypothesis which tells that there is no statistically significant effect exists in the given set of observations. It is also known as conjecture and is used in quantitative analysis to test theories about markets, investment, and finance to decide whether an idea is true or false.Null There is no difference in the mean test scores between students who study with flashcards and those who do not. 
2. Alternative Hypothesis(H1): An alternative hypothesis is a direct contradiction of the null hypothesis, which means if one of the two hypotheses is true, then the other must be false. In other words, an alternative hypothesis is a type of statistical hypothesis which tells that there is some significant effect that exists in the given set of observations.There is a difference in the mean test scores between students who study with flashcards and those who do not.
