## Seaborn : Data Visualization Library 

1. Advanced visualization library built on Matplotlib.
2. Provides attractive and informative statistical graphics.
3. Simplifies creation of complex visualizations from data.
4. It has some built-in datasets as sklearn libary

```python
# importing the libraries 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
# importing the datasets (total bill vs tip dataset)
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# visulize this data 
sns.relplot(data=tips,x ='total_bill',y='tip',col='time',hue = 'smoker',style = 'smoker',size ='size')
```

```python
sns.set_theme()
```

```python
# visulize this data 
sns.relplot(data=tips,x ='total_bill',y='tip',col='time',hue = 'smoker',style = 'smoker',size ='size')
```

```python
# iris dataset from sklearn libary 
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(iris.data,columns = iris.feature_names)
iris_df.head()
```

##### Distribution plot

```python
sns.displot(iris_df['petal width (cm)'])
```

```python
sns.distplot(iris_df['petal width (cm)'])
```

##### Correlation : 
1. Positive
2. Negative

for this we are using heat map

```python
correlation = iris_df.corr()
```

```python
print(correlation)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 10}, cmap='Blues')
```
