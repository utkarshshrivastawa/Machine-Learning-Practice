#### Manipulating a DataFrame

```python
# adding a column to a dataframe
iris_df['total'] = iris_datasets # or a both can be used
```

```python
iris_df.head()
```

```python
# removing a particular row drop( index= number) , axis = 0 for row and  1 for column
iris_df.drop( index=2 , axis = 0  )
```

```python
# for removing a particular column drop( columns= name , axis = 1)
iris_df.drop(columns = 'total' , axis= 1) 
# for permanent removing any column or row from the dataset you have to create new varible and store the table 
z = iris_df.drop(columns = 'total' , axis= 1) 
print(z)
```

```python
# locating particular using index value iloc[]row 
z.iloc[2]
```

```python
# locating particular column 
print(z.iloc[ : ,0]) # first column
print(z.iloc[:,1]) # second column 
print(z.iloc[:,2]) # third column
print(z.iloc[:,-1]) # last column
```

#### Correlation 
Correlation in pandas measures the linear relationship between columns, ranging from -1 to 1.
1. Positive Correlation : Two variables move in the same direction, indicating a direct relationship.
2. Negative Correlation :Variables move in opposite directions, indicating an inverse relationship.

```python
print(z.corr())
print(b.corr())
print(random_df.corr())
```
