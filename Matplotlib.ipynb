## Matplotlib: Useful for making plots
1. Python library for creating static, animated, and interactive visualizations.
2. Supports various plots like histograms, bar charts, and scatter plots.
3. Essential for data visualization in scientific and analytical applications.

```python
# importing matpoltlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy to get date for our plots 
import numpy as np 
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)   # sin value 
z = np.cos(x)   # cosin value 
```

```python
print(y)
```

```python
print(z)
```

```python
# creating a sin polt 
plt.plot(x,y)  # for ploting the graph 
plt.show()  # for showing the graph 
```

```python
# cosin graph 
plt.plot(x,z)
plt.show()
```

```python
# adding title , x-axix and y-axis labels 
plt.plot(x,y)
plt.xlabel('angle') # for x axis value
plt.ylabel('sine values') # for y axis value 
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('angle') # for x axis value
plt.ylabel('cosin values') # for y axis value 
plt.title('cosin wave')
plt.show()
```

```python
# parabola graph 
x= np.linspace(-10 , 10 , 20 )
y = x**2
plt.plot(x,y)
plt.xlabel('Independent value')
plt.ylabel('Dependent Value')
plt.title('Parabola Shape Curve')
plt.show()
```

```python
# Changing the way of ploting as line graph is by default
plt.plot(x,y,'r*')    # r represent the red color and * represent the graph in * marker  
plt.show()
```

```python
plt.plot(x,y,'g.')    # r represent the red color and . represent the graph . marker 
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

##### Bar Graph 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people= [100,50,150,40,70]
ax.bar(languages , people)
plt.xlabel('LANGUAGES')
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```

#####  Pie chart 

###### autopct
1. 'autopct' is an option in pie chart functions for displaying percentage format labels automatically.
2. It allows custom string formatting codes, enabling precise control over label presentation in plots.

```python
a = plt.figure()
ax = a.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people= [100,50,150,40,70]
ax.pie(people , labels = languages , autopct = '%1.1f%%')
plt.show()
```

##### Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
b = plt.figure()
ax = b.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color = 'b')
plt.show()
```

##### 3D Scatter Plot 

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,cmap = 'Blues')
plt.show()
```
